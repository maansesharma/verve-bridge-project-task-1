```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv(r"Downloads/Amazon Sales data.csv")
```

```python
plt.figure(figsize= (3,4))
sns.countplot(x = "Order Priority" , data = file)
plt.title("Count of Order priority")
plt.savefig("Count of Order priority.png")
plt.show()
```

```python
plt.figure(figsize= (2,3))
plt.show()
```

```python
count = file["Country"].value_counts
sns.countplot(x = "count" , data = file)
plt.figure(figsize = (5,6) )
plt.title(" country count")
plt.savefig(" country count.png")

plt.show()
```

```python
x = file['Item Type'].value_counts().index
y = file['Item Type'].value_counts().values
plt.pie(y,labels=x, autopct ="%0.2f%%")
plt.title("item with percentage")
plt.savefig("item with percentage")

plt.show()
```

```python
file["Region"].value_counts()
sns.countplot(x = "Region" , data = file , hue = "Item Type")
plt.figure(figsize=(5,3))
plt.title(" item count")
plt.savefig(" item count.png")
plt.show()
```

```python
file["Item Type"].value_counts()
sns.countplot(x = "Item Type" , data = file, hue = "Item Type")
plt.figure(figsize=(20,60))
plt.savefig("Item Type.png")
plt.show()
```
